```python
# 02_data_visualization.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
try:
    df = pd.read_csv('BostonHousing.csv')  # Make sure this file is in the same folder
except FileNotFoundError:
    print("Error: 'BostonHousing.csv' not found. Make sure it's in the same folder.")
    raise

# Display first few rows to check data
print("First 5 rows of the dataset:")
print(df.head())

# Check for null values
print("\nNull values in each column:")
print(df.isnull().sum())

# Convert numeric columns safely (ignore errors)
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Correlation heatmap (only numeric columns)
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(12, 9))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
